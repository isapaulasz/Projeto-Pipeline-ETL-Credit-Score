

---



```dataset_slug```: identificador do dataset (ex: 'parisrohan/credit-score-classification')

```file_path```: caminho do arquivo dentro do dataset (ex: 'train.csv')

```max_tentativas```: número de tentativas em caso de falha temporária



---


````python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import numpy as np
import pandas as pd
import time

def extrair_dados_kaggle(dataset_slug: str, file_path: str, max_tentativas: int = 3):
  tentativa = 0
  while tentativa < max_tentativas:
    try:
      df = kagglehub.dataset_load(
          KaggleDatasetAdapter.PANDAS,
          dataset_slug,
          file_path,
      )
      if df is None or df.empty:
        raise ValueError("Dataframe vazio.")
      print(f"Dados extraídos com sucesso: {len(df)} linhas, {len(df.columns)} colunas.")
      return df

    except FileNotFoundError as e:
      print(f"Arquivo não encontrado: {e}")
      raise

    except Exception as e:
      tentativa += 1
      print(f"Tentativa {tentativa}/{max_tentativas} falhou: {e}")
      if tentativa < max_tentativas:
        time.sleep(5 * tentativa)
      else:
        print("Falha após múltiplas tentativas. Abortando extração.")
        raise
````

# Criação do Dataframe

````python
df_train = extrair_dados_kaggle("parisrohan/credit-score-classification", "train.csv")
````

````python
df_test = extrair_dados_kaggle("parisrohan/credit-score-classification", "test.csv")
````

````python
df_train['origem'] = 'train'
df_test['origem'] = 'test'

# Garante que test tenha a coluna faltante, preenchida com NaN
if 'Credit_Score' not in df_test.columns:
    df_test['Credit_Score'] = pd.NA

# Reordena as colunas de test para bater com train antes de concatenar
df_test = df_test[df_train.columns]

df = pd.concat([df_train, df_test], ignore_index=True)
````

````python
df = df.rename(columns={
    "ID": "id_registro",
    "Customer_ID": "id_cliente",
    "Month": "mes",
    "Name":"nome",
    "Age":"idade",
    "SSN":"seguro_social_ssn",
    "Occupation":"profissao",
    "Annual_Income": "renda_anual",
    "Monthly_Inhand_Salary": "salario_mensal_liquido",
    "Num_Bank_Accounts": "qtd_contas_bancarias",
    "Num_Credit_Card": "qtd_cartoes_credito",
    "Interest_Rate": "taxa_juro",
    "Num_of_Loan": "qtd_emprestimos",
    "Type_of_Loan": "tipo_de_emprestimo",
    "Delay_from_due_date": "atraso_dias",
    "Num_of_Delayed_Payment": "qtd_pagamentos_atrasados",
    "Changed_Credit_Limit": "variacao_limite_credito",
    "Num_Credit_Inquiries": "qtd_consultas_credito",
    "Credit_Mix": "classificacao_credito",
    "Outstanding_Debt":"divida_pendente",
    "Credit_Utilization_Ratio":"taxa_utilizacao_credito",
    "Credit_History_Age":"tempo_historico_credito",
    "Payment_of_Min_Amount":"pagamento_valor_min",
    "Total_EMI_per_month":"emi_total_ao_mes",
    "Amount_invested_monthly": "valor_investido_mensal",
    "Payment_Behaviour":"padrao_comportamento_pagamento",
    "Monthly_Balance":"saldo_mensal",
    "Credit_Score": "score_credito"
})

df.head()
````

#Limpando os dados

Criação do dataframe para limpeza
- Limpeza de espaços e underscores nas pontas de todas as colunas tipo texto
- Substituição de qualquer variação de "nan" (case-insensitive) por NaN de verdade
- Substituição de demais casos de lixo

````python
df_limpa = df.copy()

for col in df_limpa.select_dtypes(include='object').columns:
    df_limpa[col] = df_limpa[col].astype(str).str.strip().str.strip('_')


df_limpa = df_limpa.replace(
    to_replace=r'(?i)^nan$',  # (?i) = ignora maiúscula/minúscula; ^...$ = precisa ser exatamente isso
    value=np.nan,
    regex=True
)

df_limpa = df_limpa.replace(["NA", "NM", "!@9#%8", "#F%$D@*&8", "__10000__","_", "_______", "", " "], np.nan)
````

Conversão dos tipos de dados

````python
colunas_valor_float = ['renda_anual', 'salario_mensal_liquido', 'variacao_limite_credito', 'divida_pendente', 'valor_investido_mensal', 'saldo_mensal']

for col in colunas_valor_float:
  df_limpa[col] = df_limpa[col].astype(float)
````

````python
colunas_valor_int = ['idade', 'qtd_consultas_credito', 'qtd_emprestimos', 'qtd_pagamentos_atrasados', 'qtd_contas_bancarias', 'qtd_cartoes_credito']

for col in colunas_valor_int:
  df_limpa[col] = df_limpa[col].astype('Int64')
````

Transformando outliers com valores impossíveis em nulos

````python
Q1 = df_limpa['renda_anual'].quantile(0.25)
Q3 = df_limpa['renda_anual'].quantile(0.75)
IQR = Q3 - Q1
limite_superior = Q3 + 1.5 * IQR
df_limpa.loc[df_limpa['renda_anual'] > limite_superior, 'renda_anual'] = np.nan
````

````python
colunas_valores_negativos = ['idade', 'qtd_emprestimos', 'qtd_contas_bancarias', 'qtd_pagamentos_atrasados']

for col in colunas_valores_negativos:
  qtd_negativos = (df_limpa[col] < 0).sum()
  print(f"{col}: {qtd_negativos} valores negativos encontrados")
  df_limpa.loc[df_limpa[col] < 0, col] = np.nan
````

````python
df_limpa.loc[df_limpa['idade'] > 110, 'idade'] = np.nan
````

````python
colunas_valores_acima_de_40 = ['qtd_cartoes_credito', 'taxa_juro', 'qtd_pagamentos_atrasados']

for col in colunas_valores_acima_de_40:
  df_limpa.loc[df_limpa[col] > 40, col] = np.nan
````

````python
df_limpa.loc[df_limpa['qtd_consultas_credito'] > 20, 'qtd_consultas_credito'] = np.nan
````

````python
df_limpa.loc[df_limpa['qtd_contas_bancarias'] > 15, 'qtd_contas_bancarias'] = np.nan
````

````python
df_limpa.loc[df_limpa['qtd_emprestimos'] > 10, 'qtd_emprestimos'] = np.nan
````

Arredondando valores com base no contexto

````python
colunas_monetarias = [
    'renda_anual', 'salario_mensal_liquido', 'variacao_limite_credito', 'divida_pendente', 'saldo_mensal', 'valor_investido_mensal', 'emi_total_ao_mes'
]

for col in colunas_monetarias:
    df_limpa[col] = df_limpa[col].round(2)

df_limpa['taxa_utilizacao_credito'] = df_limpa['taxa_utilizacao_credito'].round(4)
````

Convertendo a coluna ```tempo_historico_credito``` para ```tempo_historico_credito_meses```

````python
import re

def converter_para_meses(valor):
    if pd.isna(valor):
        return np.nan
    match = re.match(r'(\d+) Years and (\d+) Months', valor)
    if match:
        anos, meses = int(match.group(1)), int(match.group(2))
        return anos * 12 + meses
    return np.nan

df_limpa['tempo_historico_credito_meses'] = df_limpa['tempo_historico_credito'].apply(converter_para_meses)
df_limpa = df_limpa.drop(columns=['tempo_historico_credito'])
````

Fazendo o **One Hot Encoding** da variável ```tipo_de_emprestimo```

````python
df_limpa['tem_auto_loan'] = df_limpa['tipo_de_emprestimo'].str.contains('Auto Loan', na=False)
df_limpa['tem_credit_builder_loan'] = df_limpa['tipo_de_emprestimo'].str.contains('Credit-Builder Loan', na=False)
df_limpa['tem_debt_consolidation_loan'] = df_limpa['tipo_de_emprestimo'].str.contains('Debt Consolidation Loan', na=False)
df_limpa['tem_home_equity_loan'] = df_limpa['tipo_de_emprestimo'].str.contains('Home Equity Loan', na=False)
df_limpa['tem_mortgage_loan'] = df_limpa['tipo_de_emprestimo'].str.contains('Mortgage Loan', na=False)
df_limpa['tem_payday_loan'] = df_limpa['tipo_de_emprestimo'].str.contains('Payday Loan', na=False)
df_limpa['tem_personal_loan'] = df_limpa['tipo_de_emprestimo'].str.contains('Personal Loan', na=False)
df_limpa['tem_student_loan'] = df_limpa['tipo_de_emprestimo'].str.contains('Student Loan', na=False)
# Substitui células que são compostas inteiramente por "Not Specified" repetido (com vírgulas e "and")
df_limpa['tipo_de_emprestimo'] = df_limpa['tipo_de_emprestimo'].replace(
    to_replace=r'^(Not Specified[,\s]*(and)?[,\s]*)+$',
    value=np.nan,
    regex=True
)
df_limpa = df_limpa.drop(columns=['tipo_de_emprestimo'])
````

# Tratamento de Dados

Imputando valores

````python
colunas_puxar_mesmo_cliente = [
    'idade', 'profissao', 'seguro_social_ssn', 'nome',
    'qtd_contas_bancarias', 'qtd_cartoes_credito', 'taxa_juro'
]

for col in colunas_puxar_mesmo_cliente:
    df_limpa[col] = df_limpa.groupby('id_cliente')[col].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
    )
````

````python
colunas_imputar_mediana = [
    'renda_anual', 'salario_mensal_liquido', 'divida_pendente',
    'saldo_mensal', 'valor_investido_mensal', 'emi_total_ao_mes',
    'variacao_limite_credito', 'taxa_utilizacao_credito',
    'qtd_emprestimos', 'qtd_pagamentos_atrasados', 'qtd_consultas_credito',
    'atraso_dias', 'tempo_historico_credito_meses'
]

for col in colunas_imputar_mediana:
    df_limpa[col] = df_limpa[col].fillna(df_limpa[col].median())
````

````python
colunas_categoricas = ['classificacao_credito', 'pagamento_valor_min', 'padrao_comportamento_pagamento']

for col in colunas_categoricas:
    df_limpa[col] = df_limpa[col].fillna(df_limpa[col].mode()[0])
````

````python
df_curada = df_limpa.drop(columns=['nome', 'seguro_social_ssn'])
````
